## Set parameters

```python
import numpy as np


q = 13 # prime number
n = 10 # number of equations(under q)
X = np.random.randint(0, q, (n, n))
s = np.random.randint(0, q, (n, n)) # Private key
e = np.random.rand(n, 1) # Private key

A = X @ s
B = A + e # Public key 
```

```python
print(f"{A=}")
print()
print(f"{B=}")
print()
print(f"{s=}")
print()
print(f"{e=}")
print()
```

```python
message = np.random.randint(0, 2, (n, 1))
print(f"{message=}")
```

## Encrypted message

```python
calc_M = lambda x: (q+1)//2 if x ==1 else 0

C1 = (A @ message) % q
M = np.array([calc_M(x) for x in message]).reshape(-1, 1)
C2 = (B @ message - M) % q
```

```python
print(f"{C1=}")
print()
print(f"{C2=}")
```

## Decrypted message

```python
s @ C1 - C2
```

```python
(A - B) @ message + M
```
