```python
import numpy as np
import pandas as pd
```

```python
health = list()
with open("raw-data/健康習慣.txt", "r", encoding="utf-8") as file:
    for line in file:
        health.append(line.strip())
print(health)
health = health[1:]
NIter = len(health)
```

```python
QName = ["1週間に何回、運動をしていますか？", "1日に摂取する水の量はどのくらいですか？", "最も好きな運動は何ですか？", "日常生活で心がけている健康習慣はどんなものですか？", "睡眠時間は平均して何時間ですか？"]
QType = [int, float, list, str, float]
QRange = [(0, 7), (0, 3), ["テニス", "サッカー", "バスケ"], None, (0, 24)]
QAns = list()
```

```python
for i in range(NIter):
    EachQAns = list()
    for Q in zip(QName, QType, QRange):
        if Q[1] == int:
            EachQAns.append(np.random.randint(Q[2][0], Q[2][1]))
        elif Q[1] == float:
            EachQAns.append(np.random.uniform(Q[2][0], Q[2][1]))
        elif Q[1] == list:
            EachQAns.append(np.random.choice(Q[2]))
        elif Q[1] == str:
            EachQAns.append(health[i])
        else:
            print("Error")
    QAns.append(EachQAns)
```

```python
QAns = pd.DataFrame(QAns, columns=QName)
print(QAns)
```

```python
QAns.to_csv("raw-data/健康習慣.csv", index=False, encoding="utf-8")
QAns.to_csv("raw-data/健康習慣2.csv", index=False, encoding="utf-8")
```
